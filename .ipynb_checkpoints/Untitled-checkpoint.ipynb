```python
import matplotlib.pyplot as plt #実際にプロットする
import numpy as np #データの整理
import h5py #HDF5のライブラリ
import io
```

```python
# load HDF5 data
dat = []
with h5py.File("data/949259_test2_2025-05-12_18h55.12.918.hdf5",'r') as f:
    data = f["data_collection/events/eyetracker/BinocularEyeSampleEvent"][()]
    dat = data

values = {}
keys = ["device_time","logged_time","time","left_gaze_x","left_gaze_y","left_gaze_z",
        "right_gaze_x", "right_gaze_y", "left_gaze_z", "left_pupil_measure1", "right_pupil_measure1"]
for k in keys:
    values[k] = []
for x in dat:
    for k in keys:
        values[k].append(x[k])
```

```python
# Load session data
```

```python
plt.plot(np.arange(len(dat)),values["left_pupil_measure1"],label="left pupil")
plt.plot(np.arange(len(dat)),values["right_pupil_measure1"],label="right pupil")
plt.legend()
plt.title("pupil diameter")
plt.show()
```

```python
plt.scatter(values["left_gaze_x"],values["left_gaze_y"])
plt.scatter(values["right_gaze_x"],values["right_gaze_y"])
plt.show()
```
